# To detect the sign stop using Haar case method

```python
import cv2

import matplotlib.pyplot as pt
```

```python
img = cv2.imread('image.jpg') # read the image 

img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert the image to geryscale

img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # convert the image to rgb



stop_data = cv2.CascadeClassifier('stop_data.xml') #finding 

found = stop_data.detectMultiScale(img_grey,minSize=(20,20))

amount_found = len(found)
```

```python
if amount_found !=0:
    
    
    for (x,y,width,height) in found:
        
        cv2.rectangle(img_rgb ,(x,y),(x+height,y+width),(0,255,0),5)
```

```python
pt.subplot(1,1,1)

pt.imshow(img_rgb)
pt.show()
```
